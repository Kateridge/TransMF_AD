```python
# get labels from ADNIMERGE
import pandas as pd

# get current subject list
my_data_path = "/home/kateridge/Projects/Projects/Datasets/ADNI_SPM/PET/NIIGZ_Origin"
myfiles = glob.glob(os.path.join(my_data_path, '*'))
sub_list = []
for item in myfiles:
    sub_id = item.split('/')[-1].split('.')[0]
    sub_list.append(sub_id)

# read original ADNIMERGE
csv_all = pd.read_csv('/media/kateridge/data/Datasets/ADNI_FDG/clinical_data/ADNIMERGE.csv', usecols=['PTID', 'VISCODE', 'DX_bl', 'AGE', 'PTGENDER', 'APOE4', 'MMSE', 'DX'])
csv_ADNI = csv_all[csv_all['PTID'].isin(sub_list)]
# viscode_list = ['bl', 'm06', 'm12', 'm18', 'm24', 'm36']
# csv_ADNI = csv_ADNI[csv_all['VISCODE'].isin(viscode_list)]
# csv_ADNI_dropped = csv_ADNI.dropna(subset=['DX'])
ADNI = pd.DataFrame(columns=['Subject', 'Group', 'Age', 'Sex', 'MMSE', 'APOE'])
for sub in sub_list:
    sub_allstage = csv_ADNI[csv_ADNI['PTID'] == sub]
    sub_allstage_sorted = sub_allstage.sort_values(by='VISCODE')
    if len(sub_allstage_sorted) == 0:
        print(sub)
        continue
    # fill in the missing DX box
    if pd.isnull(sub_allstage_sorted.iloc[0, -1]):
        if sub_allstage_sorted.iloc[0, 2] == 'EMCI' or sub_allstage_sorted.iloc[0, 2] == 'EMCI':
            sub_allstage_sorted.iloc[0, -1] = 'MCI'
        elif sub_allstage_sorted.iloc[0, 2] == 'AD':
            sub_allstage_sorted.iloc[0, -1] = 'Dementia'
        elif sub_allstage_sorted.iloc[0, 2] == 'CN':
            sub_allstage_sorted.iloc[0, -1] = 'CN'
    # change visit code into numbers
    for i in range(len(sub_allstage_sorted)):
        curr_viscode = sub_allstage_sorted.iloc[i, 1]
        if curr_viscode == 'bl':
            curr_viscode = 0
        else:
            curr_viscode = int(curr_viscode.split('m')[1])
        sub_allstage_sorted.iloc[i, 1] = curr_viscode
    sub_allstage_sorted = sub_allstage_sorted.sort_values(by='VISCODE')

    sub_bl = sub_allstage_sorted.iloc[0]
    sub_allstage_nonan_sorted = sub_allstage_sorted.dropna(subset=['DX'])
    sub_allstage_nonan_sorted_list = sub_allstage_nonan_sorted.iloc[:, -1].to_list()
    sub_bf36_sorted = sub_allstage_sorted[sub_allstage_sorted['VISCODE'] <= 36]
    sub_bf36_nonan_sorted = sub_bf36_sorted.dropna(subset=['DX'])
    if sub_bl.DX == 'MCI':
        if sub_bf36_nonan_sorted.iloc[-1].DX == 'Dementia':
            ADNI.loc[len(ADNI.index)] = [sub_bl.PTID, 'pMCI', sub_bl.AGE, sub_bl.PTGENDER, sub_bl.MMSE, sub_bl.APOE4]
            continue
        else:
            if 'Dementia' in sub_allstage_nonan_sorted_list:
                print(sub_allstage_sorted)
                continue
            if sub_allstage_sorted.iloc[-1, 1] < 24:
                print(sub_allstage_sorted)
                continue
            ADNI.loc[len(ADNI.index)] = [sub_bl.PTID, 'sMCI', sub_bl.AGE, sub_bl.PTGENDER, sub_bl.MMSE, sub_bl.APOE4]
    elif sub_bl.DX == 'Dementia':
        if (not 'MCI' in sub_allstage_nonan_sorted_list) and (not 'CN' in sub_allstage_nonan_sorted_list):
            ADNI.loc[len(ADNI.index)] = [sub_bl.PTID, 'AD', sub_bl.AGE, sub_bl.PTGENDER, sub_bl.MMSE, sub_bl.APOE4]
        else:
            print(sub_allstage_sorted)
    elif sub_bl.DX == 'CN':
        if (not 'MCI' in sub_allstage_nonan_sorted_list) and (not 'Dementia' in sub_allstage_nonan_sorted_list):
            ADNI.loc[len(ADNI.index)] = [sub_bl.PTID, 'CN', sub_bl.AGE, sub_bl.PTGENDER, sub_bl.MMSE, sub_bl.APOE4]
        else:
            print(sub_allstage_sorted)
        # ADNI.append({'Subject': sub_bl.PTID, 'Group':'CN', 'Age': sub_bl.AGE, 'Sex': sub_bl.PTGENDER, 'MMSE': sub_bl.MMSE, 'APOE': sub_bl.APOE4}, ignore_index=True)
```

```python
# PET post process: intensity normalization and skulltripping according to corresponding MRI
import SimpleITK as sitk
import numpy as np
input_path ="/media/kateridge/data/Datasets/ADNI_EXTRA/Processed" # input data root
out_path="/media/kateridge/data/Datasets/ADNI_EXTRA/Final" # output data root
mri_files = glob.glob(os.path.join(input_path, 'MRI', 'w*')) # glob all MRI files

for file in mri_files:
    filename = file.split('/')[-1]
    filename = filename[1:]

    # Create Mask
    sub_t1 = sitk.ReadImage(file)
    sub_t1_np = sitk.GetArrayFromImage(sub_t1)
    sub_t1_mask = np.int8(sub_t1_np > 0)
    # maskout pet image
    sub_pet_path = file.replace('MRI', 'PET')
    sub_pet_img = sitk.ReadImage(sub_pet_path)
    sub_pet_img_np = sitk.GetArrayFromImage(sub_pet_img)
    sub_pet_img_np = sub_pet_img_np * sub_t1_mask
    # intensity normalize
    sub_pet_img_np = sub_pet_img_np.astype(np.float32)
    sub_pet_img_np /= sub_pet_img_np.mean()
    # save pet image
    sub_pet_img_out = sitk.GetImageFromArray(sub_pet_img_np)
    sub_pet_img_out.SetDirection(sub_pet_img.GetDirection())
    sub_pet_img_out.SetOrigin(sub_pet_img.GetOrigin())
    sub_pet_img_out.SetSpacing(sub_pet_img.GetSpacing())
    sitk.WriteImage(sub_pet_img_out, os.path.join(input_path, 'PET_smooth', filename))
    # save mri image
    sitk.WriteImage(sub_t1, os.path.join(out_path, 'MRI', filename))
    # print('Saved ' + sub_id_mod)
```
